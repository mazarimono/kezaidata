```python
df1999 = pd.read_csv('jp_consumer1999.csv', index_col=0)
df2004 = pd.read_csv('jp_consumer2004.csv', index_col=0)
df2009 = pd.read_csv('jp_consumer2009.csv', index_col=[0, 1], thousands=',')
df2014 = pd.read_csv('jp_consumer2014.csv', index_col=[0, 1], thousands=',')

pop2000 = pd.read_csv('2000-house.csv', thousands=',' ,index_col=0)
pop2005 = pd.read_csv('2005-house.csv', thousands=',' ,index_col=0)
pop2010 = pd.read_csv('2010-house.csv', thousands=',' ,index_col=0)
pop2015 = pd.read_csv('2015-house.csv', thousands=',' ,index_col=0)

peo2000 = pd.read_csv('2000-popu.csv', thousands=',' ,index_col=0, header=0)
peo2005 = pd.read_csv('2005-popu.csv', thousands=',' ,index_col=0, header=0)
peo2010 = pd.read_csv('2010-popu.csv', thousands=',' ,index_col=0, header=0)
peo2015 = pd.read_csv('2015-popu.csv', thousands=',' ,index_col=0, header=0)
```

```python
fig = plt.figure(figsize=(14,7))
plt.plot(df1999.iloc[3, :].astype(float), label="1999")
plt.plot(df2004.iloc[3, :].astype(float), label="2004", linestyle='--')
plt.plot(df2009.iloc[3, :].astype(float), label="2009", linestyle='-.')
plt.plot(df2014.iloc[3, :].astype(float), label="2014", linestyle=':')
plt.title('Number per household', fontsize=20)
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
dfhh = pd.concat([df1999.iloc[3, :].astype(float), df2014.iloc[3, :].astype(float)], axis=1)
dfhh.columns = ['1999', '2014']
dfhh[''] = dfhh['2014'] - dfhh['1999']
dfhh['%'] = (dfhh[''] / dfhh['1999']) * 100
fig = plt.figure(figsize=(14, 5))
plt.plot(dfhh['%'])
plt.title('', fontsize=20)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
